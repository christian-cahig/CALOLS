## <a id="37ntf-a"></a>37NTF-A

```python
import matplotlib, numpy, pandas
from matplotlib import pyplot
```

```python
# Retrieve and load dataset from spreadsheet file
dataset_workbook_path = "./37NTF-A.xlsx"
dataset_overview = pandas.read_excel(dataset_workbook_path, sheet_name="Overview")
dataset_overview
```

```python
# Containers...
# ...for consumer IDs
consumer_IDs = dataset_overview["Consumer ID"].values
# ...for total (sum of) appliance ratings per priority level among all consumers
consumer_TRPL = numpy.zeros(
    shape=(len(dataset_overview["Consumer ID"]), dataset_overview["No. of priority levels"].max())
)
# ...for appliance count per priority level among all consumers
consumer_ACPL = numpy.zeros(
    shape=(len(dataset_overview["Consumer ID"]), dataset_overview["No. of priority levels"].max())
)
# ...for priority level labels
prio_level_labels = [f"$p_c={p+1}$" for p in range(consumer_TRPL.shape[1])]
```

```python
for i in range(len(consumer_IDs)):
    dataset_consumer = pandas.read_excel(dataset_workbook_path, sheet_name=consumer_IDs[i])
    
    for j, prio_level in enumerate(dataset_consumer.columns):
        consumer_TRPL[i,j] = dataset_consumer[prio_level].dropna().sum(axis=0)
        consumer_ACPL[i,j] = len(dataset_consumer[prio_level].dropna())
```

```python
# Workspace cleanup
del i, dataset_consumer, j, prio_level
```

### <a id="37ntf-a--trpl"></a>Total appliance ratings per priority level among all consumers

```python
TRPL_df = pandas.DataFrame(
    [consumer_TRPL[i,:] for i in range(consumer_TRPL.shape[0])],
    columns = prio_level_labels
)
```

```python
ax_TRPL = TRPL_df.plot.bar(stacked=True, fontsize=7, colormap="tab10")

#ax_TRPL.set_xlabel("Consumer IDs", fontsize=7)
ax_TRPL.set_ylabel("$L_{c}$ (kW)", fontsize=7)

ax_TRPL.set_xticklabels(consumer_IDs, rotation=60)

ax_TRPL.grid(True, which="both", axis="y")
ax_TRPL.legend(loc="best", ncol=5, prop={"size":7});
```

```python
ax_TRPL.figure.savefig(
    fname="37NTF-A TRPL.png", dpi=300,
    transparent=True, bbox_inches="tight", pad_inches=0.01
)
```

### <a id="37ntf-a--acpl"></a>Total appliance counts per priority level among all consumers

```python
ACPL_df = pandas.DataFrame(
    [consumer_ACPL[i,:] for i in range(consumer_ACPL.shape[0])],
    columns = prio_level_labels
)
```

```python
ax_ACPL = ACPL_df.plot.bar(stacked=True, fontsize=7, colormap="Set3")

#ax_ACPL.set_xlabel("Consumer IDs", fontsize=7)
ax_ACPL.set_ylabel("No. of appliances", fontsize=7)

ax_ACPL.set_xticklabels(consumer_IDs, rotation=60)

ax_ACPL.grid(True, which="both", axis="y")
ax_ACPL.legend(loc="best", ncol=5, prop={"size":7});
```

```python
ax_ACPL.figure.savefig(
    fname="37NTF-A ACPL.png", dpi=300,
    transparent=True, bbox_inches="tight", pad_inches=0.01
)
```
